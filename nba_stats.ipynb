##### Visualisation

```python
from matplotlib.patches import Circle, Rectangle, Arc

def add_patches(p):
    ax=plt.gca()
    for i in p:
        ax.add_patch(i)
```

```python
def plot_shotchart(sc):
    plt.figure(figsize=(9,12))
    plt.xlim(-300,300)
    plt.ylim(-100,700)

    patches = [Rectangle((-250, -47.5), 500, 0, linewidth=2, color='black'),
    Rectangle((-250, -47.5), 0, 470, linewidth=2, color='black'),
    Rectangle((250, -47.5), 0, 470, linewidth=2, color='black'),
    Rectangle((-220, -47.5), 0, 140, linewidth=2, color='black'),
    Rectangle((220, -47.5), 0, 140, linewidth=2, color='black'),
    Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=2, color='black'),
    Rectangle((-30, -7.5), 60, 0, linewidth=4, color='black'),
    Circle((0, 0), 7.5, linewidth=2, color='black', fill=False),
    Circle((0, 142.5), 60, linewidth=2, color='black', fill=False),
    Rectangle((-60, 142.5), 120, 0, linewidth=2, color='black'),
    Rectangle((-60, -47.5), 120, 190, linewidth=2, color='black', fill=False),
    Rectangle((-80, -47.5), 160, 190, linewidth=2, color='black', fill=False)]

    add_patches(patches)
    colors=["#f75036", "#37f275"]
    labs=["FG Missed", "FG Made"]

    for i in [0,1]:
        fg=sc[sc.SHOT_MADE_FLAG==i][["LOC_X","LOC_Y"]]
        plt.scatter(-fg.LOC_X, fg.LOC_Y, s=2, c=colors[i], label=labs[i])

    #ax.tick_params(width=10, colors='r')
    plt.legend()
    plt.show()
```

```python
plot_shotchart(kobe)
```
